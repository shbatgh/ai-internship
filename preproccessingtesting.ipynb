```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import skimage.io
import skimage.morphology
import skimage
from skimage.color import rgb2gray
import scipy

image = Image.open('BBBC005_v1_images/SIMCEPImages_A24_C100_F1_s11_w2.TIF')
image = np.array(image)
#print(image.format, image.size, image.mode)
plt.imshow(image)
```

```python
import os
path = '/home/sam/dev/BBBC005_v1_images/'
images = []

for files in os.listdir(path):
    if files.endswith('.jpg'):
        images.append("../BBBC005_v1_images/" + files)
    else:
        continue
```

```python
for image in images:
    current_image = np.array(Image.open(image))
    filtered = skimage.filters.sobel(current_image)
    #plt.imshow(filtered)

    low = 0.06
    high = 0.1
    hyst = skimage.filters.apply_hysteresis_threshold(filtered, low, high)

    plt.imshow(hyst)

    labels = skimage.morphology.label(hyst)

    plt.imshow(labels)

    labeled_array, num_features = scipy.ndimage.label(labels)
    #print(num_features)
```

```python
sobel = lambda a : skimage.filters.sobel(a)
hyst = lambda a, b, c : skimage.filters.apply_hysteresis_threshold(a, b, c)
labels = lambda a : skimage.morphology.label(a)
labeled_array = lambda a : scipy.ndimage.label(a)
```

```python
def preprocessing(image):
    image = labels(hyst(image, 40, 150))

    return image
```
