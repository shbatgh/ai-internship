```python
from PIL import Image
import os

path = "cell_images/"

for file in os.listdir(path):
    im1 = Image.open(r'cell_images/' + file)
    im1.save(r'cell_images/' + file + '.png')
```

```python
for count, file in enumerate(os.listdir(path)):
    fille = "cell_images/" + file
    new_name = str(count) + '.png'
    os.rename(fille, new_name)
```

```python
for file in os.listdir(path):
    if file.endswith('.jpg'):
        print(file[:-3])
```
